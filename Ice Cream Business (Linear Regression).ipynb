```python
#importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#storing the file in the dataframe

df=pd.read_csv('IceCreamData.csv')
df.head()
```

```python
x=df['Temperature']
y=df['Revenue']
```

```python
#bar plot
fig,axes=plt.subplots(dpi=100)
axes.set_title('Analyzing IceCream Data')
axes.set_xlabel('Temperature')
axes.set_ylabel('Revenue')
axes.bar(x,y)
```

```python
#heatmap
import seaborn as sns
sns.heatmap(df.corr())
```

```python
#splitting the data into train and test set

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#fitting linear regression model to training test

from sklearn.linear_model import LinearRegression
arr1=np.array(x_train)
arr2=np.array(y_train)
x_train=arr1.reshape(-1,1)
y_train=arr2.reshape(-1,1)
model_fit=LinearRegression()
model_fit.fit(x_train,y_train)
```

```python
#predicting the future results
arr3=np.array(x_test)
x_test=arr3.reshape(-1,1)
y_pred=model_fit.predict(x_test)
```

```python
#evaluating the model, calculating the R squared value
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
